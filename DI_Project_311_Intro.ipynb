Now I know how to parse the data, the easiest way to deal with splitting the data might be to create an extra column in the dataFrame.  For this, I will create a function, which I pass it the date information, and then create a new column using the .apply() method of pandas.

```python
def YearColumn(text):
    
    if isinstance(text,str):
        time1 = parser.parse(text)
        return time1.year
    return np.nan
```

Now I will apply it to the New York City dataFrame.

```python
dfNYC["Year"] = dfNYC["Created Date"].apply(YearColumn)
```

Now, to check that the column has been added correctly, I will once again check the head and the tail of the dataframe.

```python
dfNYC.tail()
```

Yes, it has worked exactly as I expected.  Now, I can extract the data as a function of years and put them into dataframes.
